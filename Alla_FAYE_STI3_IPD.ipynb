```python
# EXERCICE 3: //-//

# 1.
print("1-")
def mediane_liste(liste_nombre):
    sort_liste = sorted(liste_nombre)
    n = len(sort_liste)
    mid = n // 2
    
    if n % 2 == 0:
        return (sort_liste[mid - 1] + sort_liste[mid]) / 2
    else:
        return sort_liste[mid]
    
nombres = [5, 2, 8, 3, 7, 1, 6, 4]
mediane_resultat = mediane_liste(nombres)
print(f"Liste de nombres {nombres}")
print(f"Mediane: {mediane_resultat}")    

print("\n")    
print("//////////////////////////////////////////////////////////////////////////////////////////") 
print("\n")
###############################################################################

# 2.
print("2-")
def filtrer_lignes(lignes):
    return [s for s in lignes if len(s) > 5]
liste_ligne = ["ananas", "banana", "orange", "mango"]
filtr_lignes = filtrer_lignes(liste_ligne)
print(f"Liste original: {liste_ligne }")
print(f"Ligne filtrer: {filtr_lignes }")

print("\n")    
print("//////////////////////////////////////////////////////////////////////////////////////////") 
print("\n")
###############################################################################

# 3. 
print("3-")
def calcul_moyenne(nombres):
    total = 0
    compte = 0
    
    for nombre in nombres:
        total += nombre
        compte += 1

    if compte == 0:
        return 0  # To avoid division by zero if the list is empty
    else:
        return total / compte
    
nombres = [10, 20, 30, 40, 50,60]
result_moyenne = calcul_moyenne(nombres)
print(f"Liste de nombres: {nombres}")
print(f"Moyenne: {result_moyenne}")

print("\n")    
print("//////////////////////////////////////////////////////////////////////////////////////////") 
print("\n")
############################################################################

# 4.
print("4-")
def liste_inverse(liste):
    return liste[::-1]

liste_original = [1, 2, 3, 4, 5]
liste_invers = liste_inverse(liste_original)
print(f"Liste original: {liste_original}")
print(f"Liste inverse: {liste_invers}")

print("\n")    
print("//////////////////////////////////////////////////////////////////////////////////////////") 
print("\n")
######################################################################

# 5.
print("5-")
# Create un liste de noms

noms = ["Alla", "Modou", "Saliou", "Fallou"]
filtrer_noms = [nom for nom in noms if 'a' in nom.lower()]
print(f"Nom original: {noms}")
print(f"Nom contenant la lettre 'a': {filtrer_noms}")
    
```

```python
#EXERCICE 4 :

# 1. 
print("1-")
def enlever_doublons(liste):
    return list(set(liste))


liste_originale = [1, 2, 3, 2, 4, 5, 6, 4]
liste_sans_doublons = enlever_doublons(liste_originale)

print("Liste originale :", liste_originale)
print("Liste sans doublons :", liste_sans_doublons)

print("\n")    
print("//////////////////////////////////////////////////////////////////////////////////////////") 
print("\n")
######################################################################

# 2.
print("2-")
def nombres_premiers(jusqu_a):
    primes = []
    for num in range(2, jusqu_a + 1):
        is_prime = all(num % i != 0 for i in range(2, int(num**0.5) + 1))
        if is_prime:
            primes.append(num)
    return primes

# Exemple d'utilisation :
limite = 20
liste_nombres_premiers = nombres_premiers(limite)

print(f"Liste de nombres premiers jusqu'à {limite} :", liste_nombres_premiers)

print("\n")    
print("//////////////////////////////////////////////////////////////////////////////////////////") 
print("\n")
######################################################################

# 3.
print("3-")
def multiplier_par_trois(nombre):
    return nombre * 3


liste_originale = [2, 4, 6, 8, 10]

# Utiliser la fonction map pour appliquer la multiplication par 3 à chaque élément
liste_multipliee = list(map(multiplier_par_trois, liste_originale))


print("Liste originale :", liste_originale)
print("Liste multipliée par 3 :", liste_multipliee)

    
```

```python
# EXERCICE 5:

def frequences_lettres(texte):
    # Initialiser le dictionnaire des fréquences
    frequences = {}

    # Convertir le texte en minuscules pour éviter la distinction entre majuscules et minuscules
    texte = texte.lower()

    # Parcourir chaque caractère dans le texte
    for char in texte:
        # Vérifier si le caractère est une lettre de l'alphabet
        if char.isalpha():
            # Ajouter le caractère au dictionnaire des fréquences
            frequences[char] = frequences.get(char, 0) + 1

    return frequences

# Exemple d'utilisation :
texte_exemple = "Bonjour, comment ça va?"
resultat_frequences = frequences_lettres(texte_exemple)

print("Fréquences des lettres dans le texte :")
for lettre, frequence in resultat_frequences.items():
    print(f"{lettre}: {frequence}")
```
